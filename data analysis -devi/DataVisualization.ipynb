```python
x=np.linspace(0,10,50)
plt.plot(x,np.sin(x),color='magenta',label='sin')
plt.plot(x,np.cos(x),color='yellow',label='cos')
plt.title("sin vs cos")
plt.xlabel("x-axis")
plt.ylabel("y-Axis")
plt.legend(loc=(1.0,1.1))
plt.show()
```

# Scatter plot

```python
x=np.array([14,22,39,42,49])
y=np.array([68,74,85,67,95])
plt.scatter(x,y,color="red",marker="v",)
plt.title("scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(10,50)
plt.ylim(50,100)
plt.show()
```

```python
x=np.array([14,22,39,42,49])
y=np.array([68,74,85,67,95])
plt.scatter(x,y,color="red",marker="v",)
plt.title("scatter plot")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.xlim(10,50)
plt.ylim(50,100)
for a,b in zip(x,y):
    plt.text(a,b,(str(a),str(b)))
plt.show()
```

# subplots:


```python
x=np.array([14,22,39,42,49])
y=np.array([68,74,85,67,95])

plt.subplot(2,2,1)
plt.plot(x,y,'g')


plt.subplot(2,2,2)
plt.plot(x,y*2,'m')


plt.subplot(2,2,3)
plt.plot(x,y*3,'y')


plt.subplot(2,2,4)
plt.plot(x*2,y,'r')


plt.show()
```

### barplot

#### vertical bargraph

```python
x=np.array(['chinnu','chinna','chinni','chitra'])
y=np.array([99,68,92,46])
plt.bar(x,y)
for a,b in zip(x,y):
    plt.text(a,b,(str(a),str(b)))
plt.show()
```

#### horizontal bargraph

```python
x=np.array(['chinnu','chinna','chinni','chitra'])
y=np.array([99,68,92,46])
plt.barh(x,y)

plt.show()
```

# Pie plot

```python
lang=['kova','conda','see','python','NET']
count=[98,100,67,98,40]
plt.pie(count,labels=lang,explode=[0,0,0,0,0.1],autopct='%1.2f%%')
plt.show()
```

```python
branch=['cse','ece','eee','it']
values=[98,79,67,90]
plt.pie(values,labels=branch,explode=[0.1,0,0,0],autopct='%1.2f%%')
plt.show()
```

```python
from sklearn.datasets import load_iris
d=load_iris()
df=pd.DataFrame(d.data,columns=d.feature_names)
df['target']=d.target
df.head(2)
```

```python
l=[]
for i in df['target']:
    if(i==0):
        l.append('satosa')
    elif(i==1):
        l.append('versicolor')
    elif(i==2):
        l.append('virginica')
l
```

```python
l=[]
for i in df['target']:
    if(i==0):
        l.append('satosa')
    elif(i==1):
        l.append('versicolor')
    elif(i==2):
        l.append('virginica')
df['classes']=l
df.sample(10)
```

```python
# catplot [cateogeorical plot]
sns.catplot(data=df)
```

```python
sns.catplot(data=df)
plt.xticks(rotation=25)   # specifies the rotation of xlimit names
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df)
plt.show() 
```

```python
sns.catplot(x='classes',y='petal length (cm)',data=df)
plt.show()
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df,kind='swarm')
plt.show()
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df,kind='box')
plt.show()
# dots far away in outp[ut are outliers]
```

### violin plot

```python
sns.violinplot(x='classes',y='sepal width (cm)',data= df)
plt.show()
```

## joint plot


It is a comination of scatter and hist

```python
sns.jointplot(x='sepal width (cm)',y='petal length (cm)',data=df)
plt.show()
```

```python
sns.jointplot(x='sepal width (cm)',y='petal length (cm)',data=df,kind='reg')
plt.show()
```

```python
sns.jointplot(x='sepal width (cm)',y='petal length (cm)',data=df,kind='hex')
plt.show()
```

### pair plot 

it is to plot multi-dimensional data

```python
sns.pairplot(data=df,hue='classes')
plt.show()
```

### correlation

It provides the relation netween two features

* relation=0--> there is no correlation between features
* relation is 1--> positive correlation
* relation is -1 --> negative correlation

```python
cor=df.corr()
cor
```

```python
sns.heatmap(cor,annot=True)
```

```python
sns.heatmap(cor,annot=True)
plt.show()
```

### Regression plot
It provides the linear model of data


it is used to Lmplot for data visualization


```python
sns.lmplot(x='sepal width (cm)',y='petal length (cm)',data=df,hue='classes')
plt.show()
```
