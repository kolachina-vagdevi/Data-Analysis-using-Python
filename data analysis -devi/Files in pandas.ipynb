```python
#concate
d1={'city':['vsp','guntur','hyd'],
    "count":[200,100,240],
    "square":[200.45,343.54,556.6]}
d1
```

```python
df1=pd.DataFrame(d1)
df1
```

```python
d2={'city':['chennai','bangalore','delhi'],
    "count":[280,190,280],
    "square":[500.45,393.54,656.6]}
d2
```

```python
df2=pd.DataFrame(d2)
df2
```

```python
total_df=pd.concat([df1,df2])
total_df
```

```python
total_df=pd.concat([df1,df2],ignore_index=True)
total_df
```

```python
total_df=pd.concat([df1,df2],keys=['a','b'])
total_df
```
